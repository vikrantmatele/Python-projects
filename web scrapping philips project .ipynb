```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://in.shop.lighting.philips.com/collections/philips-festive-sale?origin=&gad_source=1&gclid=Cj0KCQiA57G5BhDUARIsACgCYnxK4uVcNMp7-00ruzQDB0teBtGPCXzjUkhGtB3ZwTdB22Xwan51srEaAkXoEALw_wcB&gclsrc=aw.ds")
```

```python
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
name = soup.find(class_="mobile__only").get_text()
name
```

```python
name_tags = soup.select(".mobile__only")
name_tags
```

```python
a=name_tags[2].get_text()
a
```

```python
light_title=[]

for i in name_tags:
    a = i.get_text().replace("\n"," ")
    light_title.append(a)


light_title = light_title[:16]
light_title
```

```python
light_price = soup.select(".price")
light_price
```

```python
discount_price=[]

for i in light_price:
    a=i.get_text().replace("\n"," ")
    discount_price.append(a)  


discount_price = discount_price[:16]
discount_price
```

```python
model_price = soup.select(".price-old")
model_price
```

```python
old_price=[]
for i in model_price:
    a=i.get_text()
    old_price.append(a)    
old_price
```

```python
sale_percentage = soup.select(".sale-percent")
sale_percentage
```

```python
sale_per=[]

for i in sale_percentage:
    a=i.get_text()
    sale_per.append(a)

    
sale_per
```

```python
final_dict= {"title":light_title,"sale_price":discount_price,"actual_price":old_price,"percentage_off":sale_per}
final_dict
```

```python
import pandas as pd
lights = pd.DataFrame(final_dict)
lights
```

```python
lights.to_csv("lights_report_scrapping.csv",index=False)
```
